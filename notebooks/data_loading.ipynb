first let's load our data and all necessary python libraries

```python
#import python libraries

import numpy as np
import pandas as pd 
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
#from bokeh.model import ColumnDataSource
```

```python
#load the apartment building evaluation data
building_data = pd.read_csv('../data/raw/Apartment-Building-Evaluations-2023-current.csv')


#explore the data 

#building_data.head()
#building_data.info()
#building_data.describe()


#make a copy of the data
building_data_cp = building_data.copy()

#selecting relevant columns for analysis
#building_data_cp = building_data_cp['_id', 'NUMBERING OF PROPERTY', 'NEIGHBORHOOD', 'BUILDING CLASS CATEGORY']

#droping irrelevant columns
building_data_cp = building_data_cp.drop(columns=['NUMBERING OF PROPERTY', 'CLOTHING DROP BOXES', 'ACCESSORY BUILDINGS', 'WARD',
                                                  'LOBBY - WALLS AND CEILING', 'STORAGE AREAS/LOCKERS MAINT.', 'FENCING', 'CATCH BASINS / STORM DRAINAGE',
                                                  'CLOTHING DROP BOXES', 'INT. HANDRAIL / GUARD - MAINT.', 'PEST CONTROL LOG','MAINTENANCE LOG',
                                                  'ABANDONED EQUIP./DERELICT VEH.', 'GARBAGE/COMPACTOR ROOM', 'ELEVATOR COSMETICS',
                                                  'RETAINING WALLS', 
                                                  'INT. HALLWAY - WALLS / CEILING', 'INT. LOBBY / HALLWAY LIGHTING', 'CHUTE ROOMS - MAINTENANCE',
                                                  ])
#explore cleaned data
building_data_cp.info()
#building_data_cp.describe()
#check for missing values   
# check data types of columns
#print(building_data_cp.dtypes) 
#building_data_cp.isnull().sum() # check for missing values

# check the shape of the data
#print(building_data_cp.shape) 

#check for duplicates
#print(building_data_cp.duplicated().sum())
```

```python
#fill missing values of Year to 1900 
building_data_cp['YEAR BUILT'] = building_data_cp['YEAR BUILT'].fillna(1900)
building_data_cp['YEAR REGISTERED'] = building_data_cp['YEAR REGISTERED'].fillna(1900)
building_data_cp['EVALUATION COMPLETED ON'] = building_data_cp['EVALUATION COMPLETED ON'].fillna('2025-01-01')

#let's fill the missing values for proactive_building_score with current_building_eval_score
building_data_cp['PROACTIVE BUILDING SCORE'] = building_data_cp['PROACTIVE BUILDING SCORE'].fillna(building_data_cp['CURRENT BUILDING EVAL SCORE'])
```

```python
#recheck for missing values after filling
#building_data_cp.isnull().sum() 
building_data_cp.info()

#let's fill the na values with -1 for remaining missing values
building_data_cp = building_data_cp.fillna(-1)
```

```python
building_data_cp.info()
```

```python
#let's change date columns to datetime format
building_data_cp['EVALUATION COMPLETED ON'] = pd.to_datetime(building_data_cp['EVALUATION COMPLETED ON'])
```

```python
#building_data_cp.info()

#let's do some exploratory data analysis next
#
building_data_cp.head()
#let's change the float columns to integer
float_cols = building_data_cp.select_dtypes(include=['float64']).columns
building_data_cp[float_cols] = building_data_cp[float_cols].astype('int64')
```

```python
building_data_cp.info()
```

```python
#number of buildings with unique RSN 
print('The number of buildings with unique RSN is:', building_data_cp['RSN'].nunique()) 
```

```python
#let's group by year registered to see number of buildings registered each year
yearly_registrations = building_data_cp.groupby('YEAR REGISTERED').size().reset_index(name='Number of Buildings Registered')
print(yearly_registrations)
```

```python
#let's group by year built to see number of buildings built each year and that exist in the dataset
buildings_per_year = building_data_cp.groupby('YEAR BUILT').size().reset_index(name='Number of Buildings Built')
print(buildings_per_year)
```

```python
#let's group by the year evaluated to see number of evaluations done each year
evaluations_per_year = building_data_cp.groupby('YEAR EVALUATED').size().reset_index(name='Number of Evaluations Done')
print(evaluations_per_year)
```

```python
#let's group by the propoerty type to see number of buildings per property type for the 3 main property types
property_type_counts = building_data_cp['PROPERTY TYPE'].value_counts().reset_index()
property_type_counts.columns = ['PROPERTY TYPE', 'Number of Buildings']
print(property_type_counts)
```

```python
#let's count the number of buildings evaluated each year in each wardname
buildings_by_ward_and_year = building_data_cp.groupby(['WARDNAME', 'YEAR EVALUATED']).size().reset_index(name='#Apts Evaluated')
print(buildings_by_ward_and_year)
```

```python
#property type distribution across wards
property_type_by_ward = building_data_cp.groupby(['WARDNAME', 'PROPERTY TYPE']).size().reset_index(name='Number of Buildings')
print(property_type_by_ward)
```

```python
#let's classify and count the apartments current building eval score into categories with the following criteria:
# 0-40: Poor
# 41-70: Fair
# 71-90: Good
# 91-100: Excellent


def classify_building_score(score):
    if score >= 0 and score <= 40:
        return 'Poor'
    elif score >= 41 and score <= 70:
        return 'Fair'
    elif score >= 71 and score <= 90:
        return 'Good'
    elif score >= 91 and score <= 100:
        return 'Excellent'
    else:
        return 'Unknown'
    
building_data_cp['BUILDING SCORE CATEGORY'] = building_data_cp['CURRENT BUILDING EVAL SCORE'].apply(classify_building_score)
building_data_cp['BUILDING SCORE CATEGORY'].value_counts()
```

```python
#which ward has the highest number of buildings in the 'Poor' category?
print('Ward with highest number of poor buildings:\n' )
highest_poor_wards = building_data_cp[building_data_cp['BUILDING SCORE CATEGORY'] == 'Poor'].groupby('WARDNAME').size().reset_index(name='Number of Poor Buildings').sort_values(by='Number of Poor Buildings', ascending=False)
print(highest_poor_wards)
#ward with highest number of excellent buildings
print('\n \n Ward with highest number of excellent buildings:\n')
highest_excellent_wards = building_data_cp[building_data_cp['BUILDING SCORE CATEGORY'] == 'Excellent'].groupby('WARDNAME').size().reset_index(name='Number of Excellent Buildings').sort_values(by='Number of Excellent Buildings', ascending=False)
print(highest_excellent_wards)

```

```python
#let's see how many buildings have parking areas, the
parking_counts = building_data_cp.groupby(['PARKING AREAS', 'PROPERTY TYPE']).size().reset_index(name='Number of Buildings') 
print(parking_counts)
```

```python
#export cleaned data to csv 
building_data_cp.to_csv('../data/processed/rentsafeto_cleaned.csv', index=False)
```

```python
#let's visualize some of the data using bokeh 

output_notebook()


#plot number of buildings built each year
p1 = figure(title='Number of Buildings Built Each Year', x_axis_label='Year Built', y_axis_label='Number of Buildings', width=800, height=400)
p1.vbar(x=buildings_per_year['YEAR BUILT'], top=buildings_per_year['Number of Buildings Built'], width=0.8)
show(p1)
```

```python
#let's visualize some of the data using bokeh gridplot

#the gridplot should contain multiple plots in a grid layout 
# number of buildings built each year, 
plot1 = figure(title='Number of Buildings Built Each Year', x_axis_label='Year Built', y_axis_label='Number of Buildings', width=400, height=300)
plot1.vbar(x=buildings_per_year['YEAR BUILT'], top=buildings_per_year['Number of Buildings Built'], width=0.8)
# number of buildings registered each year
plot2 = figure(title='Number of Buildings Registered Each Year', x_axis_label='Year Registered', y_axis_label='Number of Buildings', width=400, height=300)
plot2.vbar(x=yearly_registrations['YEAR REGISTERED'], top=yearly_registrations['Number of Buildings Registered'], width=0.8)
# number of evaluations done each year 
plot3 = figure(title='Number of Evaluations Done Each Year', x_axis_label='Year Evaluated', y_axis_label='Number of Evaluations', width=400, height=300)
plot3.vbar(x=evaluations_per_year['YEAR EVALUATED'], top=evaluations_per_year['Number of Evaluations Done'], width=0.8)
# percentage of buildings per property type(piechart)
from math import pi 
from bokeh.transform import cumsum
property_type_counts['angle'] = property_type_counts['Number of Buildings']/property_type_counts['Number of Buildings'].sum() * 2 * pi
property_type_counts['color'] = ['#718dbf', '#e84d60', '#c9d9d3']
plot4 = figure(title='Percentage of Buildings per Property Type', height=350, width=400, toolbar_location=None,
               tools="hover", tooltips="@{PROPERTY TYPE}: @Number of Buildings", x_range=(-0.5, 1.0))
plot4.wedge(x=0, y=1, radius=0.4, 
            start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),  
            line_color='white', fill_color='color', legend_field='PROPERTY TYPE', source=property_type_counts)
# ward with highest number of poor buildings
# (This can be represented as a bar chart of wards vs number of poor buildings) 
poor_buildings_by_ward = building_data_cp[building_data_cp['BUILDING SCORE CATEGORY'] == 'Poor'].groupby('WARDNAME').size().reset_index(name='Number of Poor Buildings').sort_values(by='Number of Poor Buildings', ascending=False)
plot5 = figure(title='Number of Poor Buildings by Ward', x_axis_label='Ward Name', y_axis_label='Number of Poor Buildings', x_range=poor_buildings_by_ward['WARDNAME'], width=800, height=400)
plot5.vbar(x=poor_buildings_by_ward['WARDNAME'], top=poor_buildings_by_ward['Number of Poor Buildings'], width=0.8)
plot5.xaxis.major_label_orientation = pi/4  
# ward with highest number of excellent buildings
excellent_buildings_by_ward = building_data_cp[building_data_cp['BUILDING SCORE CATEGORY'] == 'Excellent'].groupby('WARDNAME').size().reset_index(name='Number of Excellent Buildings').sort_values(by='Number of Excellent Buildings', ascending=False)
plot6 = figure(title='Number of Excellent Buildings by Ward', x_axis_label='Ward Name', y_axis_label='Number of Excellent Buildings', x_range=excellent_buildings_by_ward['WARDNAME'], width=800, height=400)
plot6.vbar(x=excellent_buildings_by_ward['WARDNAME'], top=excellent_buildings_by_ward['Number of Excellent Buildings'], width=0.8)
plot6.xaxis.major_label_orientation = pi/4  
# parking area distribution
parking_counts_plot = building_data_cp.groupby('PARKING AREAS').size().reset_index(name='Number of Buildings')
# convert parking categories to strings for categorical x_range (Bokeh expects categorical values as strings)
parking_cats = parking_counts_plot['PARKING AREAS'].astype(str).tolist()
plot7 = figure(title='Parking Area Distribution', x_axis_label='Parking Areas', y_axis_label='Number of Buildings', x_range=parking_cats, width=400, height=300)
plot7.vbar(x=parking_counts_plot['PARKING AREAS'].astype(str).tolist(), top=parking_counts_plot['Number of Buildings'], width=0.8)
plot7.xaxis.major_label_orientation = pi/4
from bokeh.layouts import gridplot
grid = gridplot([[plot1, plot2], [plot3, plot4], [plot5, plot6], [plot7, None]])
show(grid)


# --- IGNORE ---    
```
